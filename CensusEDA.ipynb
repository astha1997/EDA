```python
#Checking for wrong entries like symbols -,?,#,*,etc.
for col in data :
    print(f" {col} : {data[col].unique()} ")
```

```python
data = data.rename(columns={'native.country': 'native_country'})
data = data.rename(columns={'education.num': 'class'})
data = data.rename(columns={'marital.status': 'marital_status'})
data = data.rename(columns={'capital.loss': 'capital_loss'})
data = data.rename(columns={'capital.gain': 'capital_gain'})
data = data.rename(columns={'hours.per.week': 'hours_per_week'})


```

```python
data.describe(include='all')
```

```python
import numpy as np
for col in data:
    data[col].replace({'?':np.nan},inplace=True)
```

```python
data.isnull().sum()
```

```python
workclass_mode = data.workclass.mode()[0]
```

```python
workclass_mode
```

```python
occupation_mode = data.occupation.mode()[0]
```

```python
occupation_mode
```

```python
country_mode = data['native_country'].mode()[0]
```

```python
country_mode
```

```python
data.workclass.fillna(workclass_mode, inplace = True)
data.occupation.fillna(occupation_mode, inplace = True)
data.native_country.fillna(country_mode, inplace = True)
```

```python
for col in data :
    print(f" {col} : {data[col].unique()} ")
```

```python
import seaborn as sns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'relationship', 'occupation','workclass','marital_status']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=data[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```
